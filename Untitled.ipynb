```python
def clean_text(text): 
    return text.lower()
```

```python
df = pd.read_csv('original 2.csv', encoding='cp1252')
df=pd.DataFrame(df.values)
df.rename(columns = {0:'IDUPA',1:'EventDate',2:'Hospitalized',3:'Amputation',4:'Nature',5:'NatureTitle',6:'Part of Body',7:'Part of Body Title',8:'Event',9:'EventTitle',10:'Source',11:'SourceTitle'}, inplace = True)
```

```python
tit_ind = df[df['Part of Body Title']=='Nonclassifiable' ].index
len(tit_ind)
df.drop(tit_ind , inplace=True)
df["Part of Body Title"]=df["Part of Body Title"].str.replace(r'[^\w\s]+', '')
df["Part of Body Title"]=df["Part of Body Title"].str.replace('unspecified', '')
df["Part of Body Title"] = df["Part of Body Title"].apply(clean_text)



tit_ind = df[df['EventTitle']=='Nonclassifiable' ].index
len(tit_ind)
df.drop(tit_ind , inplace=True)
df["EventTitle"]=df["EventTitle"].str.replace(r'[^\w\s]+', '')
df["EventTitle"]=df["EventTitle"].str.replace('unspecified', '')
df["EventTitle"] = df["EventTitle"].apply(clean_text)


tit_ind = df[df['NatureTitle']=='Nonclassifiable' ].index
len(tit_ind)
df.drop(tit_ind , inplace=True)
df["NatureTitle"]=df["NatureTitle"].str.replace(r'[^\w\s]+', '')
df["NatureTitle"]=df["NatureTitle"].str.replace('unspecified', '')
df["NatureTitle"] = df["NatureTitle"].apply(clean_text)


tit_ind = df[df['SourceTitle']=='Nonclassifiable' ].index
len(tit_ind)
df.drop(tit_ind , inplace=True)
df["SourceTitle"]=df["SourceTitle"].str.replace(r'[^\w\s]+', '')
df["SourceTitle"]=df["SourceTitle"].str.replace('unspecified', '')
df["SourceTitle"] = df["SourceTitle"].apply(clean_text)
```

```python
def lemmatization(texts,allowed_postags=['NOUN', 'ADJ']): 
       output = []
       for sent in texts:
             doc = nlp(sent) 
             output.append([token.lemma_ for token in doc if token.pos_ in allowed_postags ])
       return output
```

```python

text_list=df['Part of Body Title'].tolist()
print(text_list[1])
tokenized_reviews = lemmatization(text_list)
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
df.head(5)
```

```python
df2=pd.DataFrame()
df2=df[['NatureTitle','Part of Body Title','EventTitle','SourceTitle']]
```

```python
df2
```

```python
x=list(df2.values)
type(x)
```

```python
te=TransactionEncoder()
te_ary=te.fit(df2.values).transform(df2.values)
```

```python
df3=pd.DataFrame(te_ary,columns=te.columns_)
df3
```

```python
from mlxtend.frequent_patterns import apriori
frequent_itemsets=apriori(df3,min_support=0.1,use_colnames=True)
frequent_itemsets
```

```python
# Associaltive rule mining
```

```python
from mlxtend.frequent_patterns import association_rules
res = association_rules(frequent_itemsets,metric="confidence",min_threshold=0.6)
res
```

```python
res1=res[['antecedents','consequents','support','confidence','lift']]
res1
```

```python
res2=res1[res1['confidence']>=1]
res2
```
